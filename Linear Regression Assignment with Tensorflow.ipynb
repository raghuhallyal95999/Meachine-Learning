```python
#importing necessory modules
import numpy as np
import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt
tf.disable_v2_behavior()
```

```python
#mounting drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
np.random.seed(101)
tf.set_random_seed(101)
```

```python
# Generating random linear data
x = np.linspace(0, 50, 50)
y = np.linspace(0, 50, 50)
# Adding noise to the random linear data
x += np.random.uniform(-4, 4, 50)
y += np.random.uniform(-4, 4, 50)
n = len(x) # Number of data points
```

```python
# print(x)
x = []
y = []
print('extracting x-points and y-points')
with open('/content/drive/MyDrive/9.txt',mode='r',encoding='utf-8') as f:
	lines = f.read().split('\n')
	
for i in lines:
	x1,y1 = i.split(',')
	x.append(x1)
	y.append(y1)
# dataset
x = list(map(float,x))
y=list(map(float,y))
print('Doneextracting x-points and y-points')
```

```python
print(y)
```

```python
# Plot of Training Data
plt.scatter(x, y)
plt.xlabel('x-points')
plt.xlabel('y-points')
plt.title("Given Data(9.txt)")
plt.show()
```

```python
# tf.compat.v1.disable_v2_behavior()
X = tf.placeholder("float")
Y = tf.placeholder("float")
```

```python
W = tf.Variable(0., name = "W")
b = tf.Variable(0., name = "b")
```

```python
rate = 0.01
epochss = 1000
```

```python
# Hypothesis
y_pred = tf.add(tf.multiply(X, W), b)

# Mean Squared Error Cost Function
cost = tf.reduce_sum(tf.pow(y_pred-Y, 2)) / (2 * n)

# Gradient Descent Optimizer
optimizer = tf.train.GradientDescentOptimizer(rate).minimize(cost)

# Global Variables Initializer
init = tf.global_variables_initializer()
```

```python

# Starting the Tensorflow Session
EPOCHS=[]
losz=[]
theta0 = []
theta1 = []
with tf.Session() as sess:
	
	# Initializing the Variables
	sess.run(init)
	
	# Iterating through all the epochs
	for epoch in range(epochss):
		
		# Feeding each data point into the optimizer using Feed Dictionary
		for (_x, _y) in zip(x, y):
			sess.run(optimizer, feed_dict = {X : _x, Y : _y})
		
		# Displaying the result after every 50 epochs
		if (epoch + 1) % 50 == 0:
			# Calculating the cost a every epoch
			c = sess.run(cost, feed_dict = {X : x, Y : y})
			EPOCHS.append(epoch+1)
			losz.append(c)
			print("Epoch", (epoch + 1), ": cost =", c, "W =", sess.run(W), "b =", sess.run(b))
			theta0.append(sess.run(W))
			theta1.append(sess.run(b))
	# Storing necessary values to be used outside the Session
	train_cost = sess.run(cost, feed_dict ={X: x, Y: y})
	weight = sess.run(W)
	bias = sess.run(b)
```

```python
# Calculating the predictions
predictions = []
for i in x:
    predictions.append(weight*i+bias)
print("Training cost =", train_cost, "Weight =", weight, "bias =", bias, '\n')
```

```python
# Plotting the Results
plt.plot(x, y, 'o', label ='Original data')
plt.plot( x,predictions, 'r',label ='Fitted line')
plt.title('Linear Regression Result')
# plt.legend()
plt.show()
```

## Gradient descent algorithm should stop running when an error 
## (cost) is not getting decreased much compared to last few epochs 
## instead of running till last epoch  

```python
EPOCHS=[]
losz=[]
theta0 = []
theta1 = []
with tf.Session() as sess:
	
	# Initializing the Variables
	sess.run(init)
	
	# Iterating through all the epochs
	for epoch in range(epochss):

		for(_x,_y) in zip(x,y):
			sess.run(optimizer,feed_dict={X:_x,Y:_y})

		c = sess.run(cost,feed_dict={X:x,Y:y})
		EPOCHS.append(epoch+1)
		losz.append(c)
		print("Epoch : ",(epoch+1)," cost : ",c , " W : ",sess.run(W)," b : ",sess.run(b))
		theta0.append(sess.run(W))
		theta1.append(sess.run(b))
		if len(losz)>1 and losz[len(losz)-2]==c:
			break     
	# Storing necessary values to be used outside the Session
	train_cost = sess.run(cost, feed_dict ={X: x, Y: y})
	weight = sess.run(W)
	bias = sess.run(b)
```

```python

while(rate<0.1):
	# Hypothesis
	y_pred = tf.add(tf.multiply(X, W), b)

	# Mean Squared Error Cost Function
	cost = tf.reduce_sum(tf.pow(y_pred-Y, 2)) / (2 * n)

	# Gradient Descent Optimizer
	optimizer = tf.train.GradientDescentOptimizer(rate).minimize(cost)

	# Global Variables Initializer
	init = tf.global_variables_initializer()
	EPOCHS=[]
	losz=[]
	theta0 = []
	theta1 = []
	with tf.Session() as sess:
		
		# Initializing the Variables
		sess.run(init)
		
		# Iterating through all the epochs
		for epoch in range(epochss):

			for(_x,_y) in zip(x,y):
				sess.run(optimizer,feed_dict={X:_x,Y:_y})

			c = sess.run(cost,feed_dict={X:x,Y:y})
			EPOCHS.append(epoch+1)
			losz.append(c)
			print("Epoch : ",(epoch+1)," cost : ",c , " W : ",sess.run(W)," b : ",sess.run(b))
			theta0.append(sess.run(W))
			theta1.append(sess.run(b))
			if len(losz)>1 and losz[len(losz)-2]==c:
				break
		train_cost = sess.run(cost, feed_dict ={X: x, Y: y})
		weight = sess.run(W)
		bias = sess.run(b)
	real_loss.append(c)
	learn.append(rate)
	rate=rate+0.01
```
